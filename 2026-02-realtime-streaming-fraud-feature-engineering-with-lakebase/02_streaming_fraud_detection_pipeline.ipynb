This notebook demonstrates an end-to-end [Real-Time streaming](https://www.databricks.com/blog/introducing-real-time-mode-apache-sparktm-structured-streaming) feature engineering pipeline that computes **stateless and stateful features** from credit card transaction data and publishes them to a [Lakebase](https://docs.databricks.com/en/lakebase/index.html) PostgreSQL table with sub-second latency.

The pipeline reads JSON events from Kafka, applies row-level feature transformations, computes per-user stateful fraud indicators using Spark's `transformWithState` API (Spark 4.0+), and writes the enriched features to Lakebase PostgreSQL via `ForeachWriter`.

### Feature Summary

| Category | Type | Features |
|---|---|---|
| **Time-based** | Stateless | Hour of day, day of week, business hours flag, cyclical sin/cos |
| **Amount-based** | Stateless | Log transform, sqrt, amount categories, z-scores |
| **Merchant** | Stateless | Risk scores by merchant category |
| **Location** | Stateless | Risk indicators, region classification |
| **Device & Network** | Stateless | Device type detection, IP address classification |
| **Velocity** | Stateful | Transaction counts in 10-min and 1-hour windows |
| **IP Tracking** | Stateful | IP change detection, distinct IP counts |
| **Travel Anomaly** | Stateful | Impossible travel detection (velocity > 800 km/h) |
| **Amount Anomaly** | Stateful | Z-score vs rolling user history |
| **Fraud Score** | Stateful | Composite score (0-100) with binary prediction flag |

## 1: Configure Spark for Real-Time Streaming

Set up Spark configurations for stateful streaming:
- **RocksDB** state store provider for efficient state management
- **Changelog checkpointing** for faster recovery
- **Shuffle partitions** set to 4 (tune based on cluster size)

```python
#Retsart Python env to clear config and code caches.
dbutils.library.restartPython()

#Setup Spark streaming configs
spark.conf.set("spark.sql.streaming.stateStore.providerClass", "com.databricks.sql.streaming.state.RocksDBStateStoreProvider")
spark.conf.set("spark.databricks.streaming.statefulOperator.asyncCheckpoint.enabled", "false")
spark.conf.set("spark.sql.streaming.stateStore.rocksdb.changelogCheckpointing.enabled", "true")

#Adjust Shuffle partitions based on your need.
spark.conf.set("spark.sql.shuffle.partitions", "4")
```


## 2: Import Libraries

Load all required modules:
- **PySpark** functions and types
- **Config** for pipeline settings
- **AdvancedFeatureEngineering** for stateless feature computation
- **FraudDetectionFeaturesProcessor** for stateful `transformWithState` logic
- **LakebaseClient** for PostgreSQL writes

```python
# Import required libraries
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
import logging

# Import utility modules
from utils.config import Config
from utils.feature_engineering import (
    AdvancedFeatureEngineering,
    FraudDetectionFeaturesProcessor,
    get_fraud_detection_output_schema
)
from utils.lakebase_client import LakebaseClient

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```


## 3: Initialize Configuration and Verify Lakebase


Initialize all pipeline components and verify the Lakebase `transaction_features` table exists (created by `00_setup`).

<!-- <img src="./images/streaming_fraud_init_verify.png" alt="Init and Verify Flow" width="100%" style="display: block; margin: 0 auto 20px auto"/> -->

```python
#Initialize Config
config = Config()

# Initialize components
feature_engineer = AdvancedFeatureEngineering(spark)
lakebase = LakebaseClient(**config.lakebase_config)

# Test Lakebase connection
if lakebase.test_connection():
    print("Connected to Lakebase PostgreSQL")
else:
    raise Exception("Failed to connect to Lakebase")

# Verify transaction_features table exists
print("\nVerifying transaction_features table...")
try:
    stats = lakebase.get_table_stats("transaction_features")
    print(f"  Table exists with {stats['total_rows']:,} rows")
except Exception as e:
    print("  Table not found. Please run 00_setup.ipynb first!")
    raise Exception("transaction_features table does not exist")
```


## 4: Read Streaming Data from Kafka


Connect to the Kafka topic, deserialize JSON messages, and produce a structured streaming DataFrame with typed columns. Uses `startingOffsets: latest` to consume only new messages.

<img src="./images/streaming_fraud_kafka_read.png" alt="Kafka Read Flow" width="100%" style="display: block; margin: 0 auto 20px auto"/>

```python
# Retrieve Kafka credentials from Databricks secrets
kafka_credentials_secrets = config.kafka_config["kafka_credentials_secrets"]
scope = kafka_credentials_secrets["scope"]

# Retrieve secrets from Databricks secret scope.
# Note: Daatabricks secrets should be stored prior to using them and it's not covered in this example
# You can find more information about Databricks secrets here: https://docs.databricks.com/aws/en/security/secrets/#secrets-overview
KAFKA_USERNAME = dbutils.secrets.get(scope = scope, key = kafka_credentials_secrets["username"])
KAFKA_SECRET = dbutils.secrets.get(scope = scope, key = kafka_credentials_secrets["secret"])
KAFKA_SERVER = dbutils.secrets.get(scope = scope, key = kafka_credentials_secrets["server"])
KAFKA_TOPIC = config.kafka_config["kafka_topic"]

# Define the schema for the Kafka value column

value_schema = "STRUCT<transaction_id: STRING, user_id: STRING, merchant_id: STRING, amount: DOUBLE, currency: STRING, merchant_category: STRING, payment_method: STRING, ip_address: STRING, device_id: STRING, latitude: DOUBLE, longitude: DOUBLE, card_type: STRING, timestamp: TIMESTAMP>"

# Read streaming data from Kafka topic with SASL_SSL authentication
kafka_df = (
    spark
    .readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", KAFKA_SERVER)
    .option("kafka.security.protocol", "SASL_SSL")
    .option("kafka.sasl.jaas.config", f"kafkashaded.org.apache.kafka.common.security.plain.PlainLoginModule required username='{KAFKA_USERNAME}' password='{KAFKA_SECRET}';")
    .option("kafka.ssl.endpoint.identification.algorithm", "https")
    .option("kafka.sasl.mechanism", "PLAIN")
    .option("subscribe", KAFKA_TOPIC)
    .option("startingOffsets", "latest")
    .option("failOnDataLoss", "false")
    .load()
) \
.select(
    from_json(col("value").cast("string"),
              value_schema).alias("value"),
    col("timestamp").alias("kafka_timestamp")) \
.selectExpr("value.*", "kafka_timestamp") \
  .drop("timestamp") \
  .withColumnRenamed("kafka_timestamp", "timestamp")
```


## 5: Apply Stateless Features


Apply row-level feature transformations that do not require state. These are computed independently for each transaction:

- **Time-based**: hour of day, day of week, business hours flag, cyclical sin/cos encodings
- **Amount-based**: log transform, sqrt, amount categories, z-scores
- **Merchant**: risk scores derived from merchant category
- **Location**: risk indicators, region classification
- **Device**: device type detection from device_id
- **Network**: IP address classification

> 💡 Stateless features are pure row-level transformations — they have no dependency on previous events and can be computed in a single `select` pass over the streaming DataFrame.

<img src="./images/streaming_fraud_stateless_features.png" alt="Stateless Features" width="100%" style="display: block; margin: 0 auto 20px auto"/>

```python
df_with_stateless_features = feature_engineer.apply_all_features(kafka_df)
```


## 6: Apply Stateful Features with `transformWithState`


Use Spark's [`transformWithState`](https://spark.apache.org/docs/latest/structured-streaming-programming-guide.html) API (Spark 4.0+) to maintain per-user state and detect fraud patterns. The `FraudDetectionFeaturesProcessor` groups by `user_id`, applies a 1-minute watermark, and emits enriched rows in `Update` output mode.

| Feature | Description |
|---|---|
| **Velocity** | Transaction counts in 10-min and 1-hour sliding windows |
| **IP Tracking** | Detect IP address changes and count distinct IPs |
| **Location** | Impossible travel detection (velocity > 800 km/h) |
| **Amount** | Z-score calculation against rolling user history |
| **Fraud Score** | Composite score (0-100) with binary prediction flag |

> 💡 The `transformWithState` API provides exactly-once semantics with RocksDB-backed state. Each user's transaction history is maintained across micro-batches, enabling complex temporal patterns.

<img src="./images/streaming_fraud_stateful_features.png" alt="Stateful Features" width="100%" style="display: block; margin: 0 auto 20px auto"/>

```python
# Apply stateful fraud detection using transformWithState
df_with_fraud_features = df_with_stateless_features \
    .groupBy("user_id") \
    .transformWithState(
        statefulProcessor=FraudDetectionFeaturesProcessor(),
        outputStructType=get_fraud_detection_output_schema(),
        outputMode="update",
        timeMode="processingTime"
    )
```


## 7: Write to Lakebase PostgreSQL via ForeachWriter


Stream all enriched features to Lakebase PostgreSQL for real-time serving. The `ForeachWriter` batches upserts for efficiency, and the `realTime` trigger mode ensures sub-second end-to-end latency.

<!-- <img src="./images/streaming_fraud_lakebase_write.png" alt="Lakebase Write Flow" width="100%" style="display: block; margin: 0 auto 20px auto"/> -->

```python
CHECKPOINT_BASE_PATH = config.kafka_config["checkpoint_base_path"]
FRAUD_PIPELINE_CHECKPOINT_LOCATION = f"{CHECKPOINT_BASE_PATH}/fraud-pipeline-checkpoint"
dbutils.fs.rm(FRAUD_PIPELINE_CHECKPOINT_LOCATION, True)
```

```python
#Get Schema from dataframe
table_schema = df_with_fraud_features.schema

#Initialize lakebase writer
lakebase_writer  = lakebase.get_foreach_writer(column_names=table_schema.names, batch_size=5)

#Start streaming query
query = df_with_fraud_features \
    .writeStream \
    .outputMode("update") \
    .foreach(lakebase_writer) \
    .option("checkpointLocation", FRAUD_PIPELINE_CHECKPOINT_LOCATION) \
    .trigger(realTime="5 minutes") \
    .start()
```


## Summary

This pipeline demonstrates a complete real-time feature engineering workflow:

1. **Kafka ingestion** — Read JSON credit card transactions via SASL_SSL
2. **Stateless features** — Compute time, amount, merchant, location, device, and network features
3. **Stateful features** — Maintain per-user state for velocity, IP tracking, travel anomaly, amount anomaly, and composite fraud scoring
4. **Lakebase write** — Upsert enriched features to PostgreSQL with ~400ms end-to-end latency

All features are available in the `transaction_features` table for downstream model training or real-time serving via Lakebase.
